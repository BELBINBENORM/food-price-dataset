## 📖 About This Dataset

<div style="border-left:4px solid #22c55e; padding:12px 20px; background:#0d200d22;
            border-radius:0 10px 10px 0; margin:16px 0;">
  <strong style="color:#86efac;">What is this?</strong><br>
  <span style="color:#bbf7d0; font-size:14px; line-height:1.8;">
    A longitudinal macro dataset tracking global food price inflation and hunger from 1990 to today.
    Five data streams are pulled from FAO and World Bank, harmonised to monthly or annual frequency,
    and merged into a unified panel that captures how food commodity shocks, fertiliser cost surges,
    crop yield trends, and supply-chain disruptions translate into hunger and malnutrition worldwide —
    with direct cross-reference potential to tariff and trade datasets.<br><br>
    • <strong>FAO Food Price Index</strong>: Monthly FFPI overall + five sub-indices — cereals, dairy, meat, vegetable oils, sugar (Jan 1990)<br>
    • <strong>Commodity Spot Prices</strong>: Monthly wheat (US HRW), rice (Thailand 25%), maize (US #2 Yellow), soybean oil spot prices via World Bank Pink Sheet (1960)<br>
    • <strong>Fertiliser Prices</strong>: Monthly DAP (diammonium phosphate), Urea (bulk), Potassium Chloride (MOP) spot prices — the 3 primary crop input costs (1960)<br>
    • <strong>Crop Yields</strong>: Annual global area harvested, production quantity, and yield (kg/ha) for wheat, rice, and maize from FAOSTAT (1961)<br>
    • <strong>Food Security</strong>: Annual prevalence of undernourishment (% population), food production index, cereal yield index, number of undernourished people (1992)<br>
    • <strong>Derived Ratios</strong>: Fertiliser-to-food price ratio, real food price index, price volatility index, food vs income affordability gap<br>
    • <strong>ML Features</strong>: Lag-1 to Lag-6, rolling 3/6/12-month windows, YoY change rates, crisis flags, next-month and next-year targets
  </span>
</div>

## 📐 Derived Food Security Metrics

<div style="border: 1px solid #22c55e; background: #0d200d22; border-radius: 10px; padding: 16px 20px; margin: 12px 0;">
  <strong style="color:#86efac;">🌾 How food security ratios are computed</strong><br>
  <span style="color:#bbf7d0; font-size:14px; line-height:1.8;">
    Six ratios are computed on the merged monthly panel:<br><br>
    • <strong>Real Food Price Index</strong>: FAO FFPI nominal &divide; CPI all items (World Bank) &times; 100 — inflation-adjusted food price level relative to 2014–2016 = 100<br>
    • <strong>Fertiliser-to-Cereal Price Ratio</strong>: DAP price per tonne &divide; Wheat price per tonne — input cost multiplier (e.g. 3× = farmer margin squeeze)<br>
    • <strong>Food Price Volatility (12m rolling σ)</strong>: 12-month rolling standard deviation of month-over-month FFPI change — captures shock regimes<br>
    • <strong>Food vs CPI Gap</strong>: FFPI YoY% − Headline CPI YoY% — measures how much food inflation exceeds general inflation (poor households feel this first)<br>
    • <strong>Cereals Affordability Index</strong>: (100 &divide; cereals sub-index) &times; 100 — rebased so 100 = 2014–2016 baseline; &lt;100 = more expensive than baseline<br>
    • <strong>Supply Stress Flag</strong>: Binary — 1 when cereals price YoY &gt; 20% AND fertiliser price YoY &gt; 20% simultaneously (compound supply shock regime)
  </span>
</div>

```python
# ── Date range ─────────────────────────────────────────────────────────────
START_YEAR = 1990
END_DATE   = datetime.today().strftime('%Y-%m-%d')
END_YEAR   = datetime.today().year

# ── Output dir ─────────────────────────────────────────────────────────────
OUTPUT_DIR = '.'
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f'📅 Date range  : {START_YEAR} → {END_DATE}')
print(f'📂 Output dir  : {os.path.abspath(OUTPUT_DIR)}')
print('\nData sources (no API keys required):')
print('  🌿 FAO FAOSTAT — bulk download ZIPs (free, open)')
print('  🌍 World Bank Pink Sheet — monthly commodity prices (free, open)')
print('  📊 World Bank API — annual food security indicators (free, open)')
```

```python
# ── Install curl_cffi (bypasses Cloudflare TLS fingerprint blocking) ──────
import subprocess, sys
subprocess.run([sys.executable, "-m", "pip", "install", "curl_cffi", "-q"], check=True)

from curl_cffi import requests          # drop-in replacement for requests
import pandas as pd, numpy as np
import time, os, io, zipfile, warnings
from datetime import datetime
warnings.filterwarnings("ignore")

START_YEAR = 1990
END_DATE   = datetime.today().strftime("%Y-%m-%d")
END_YEAR   = datetime.today().year
OUTPUT_DIR = "."
os.makedirs(OUTPUT_DIR, exist_ok=True)

# impersonate="chrome" makes curl_cffi mimic Chrome's exact TLS handshake
# — this is what bypasses Cloudflare 403 / 521 errors on FAO / World Bank
IMPERSONATE = "chrome"

# ── FAOSTAT bulk-ZIP helper ────────────────────────────────────────────────
FAOSTAT_BASE  = "https://bulks-faostat.fao.org/production"
FAOSTAT_BASE2 = "https://fenixservices.fao.org/faostat/static/bulkdownloads"

def fetch_faostat_zip(filename, target_csv_prefix=None):
    for base in [FAOSTAT_BASE, FAOSTAT_BASE2]:
        url = f"{base}/{filename}"
        print(f"  Downloading {url} ...")
        try:
            r = requests.get(url, impersonate=IMPERSONATE, timeout=180)
            if r.status_code == 200:
                z = zipfile.ZipFile(io.BytesIO(r.content))
                csv_files = [n for n in z.namelist() if n.endswith(".csv")]
                if target_csv_prefix:
                    csv_files = [n for n in csv_files if n.startswith(target_csv_prefix)] or csv_files
                if not csv_files:
                    print(f"  No CSV found in {filename}")
                    return pd.DataFrame()
                with z.open(csv_files[0]) as fh:
                    df = pd.read_csv(fh, encoding="latin-1", low_memory=False)
                print(f"  \u2705  {csv_files[0]}  \u2192  {df.shape[0]:,} rows")
                return df
            print(f"  \u26a0\ufe0f  {r.status_code} \u2014 trying fallback URL...")
        except Exception as e:
            print(f"  \u26a0\ufe0f  {e} \u2014 trying fallback URL...")
    print(f"  \u274c  Both FAOSTAT URLs failed for {filename}")
    return pd.DataFrame()


# ── World Bank JSON API helper ─────────────────────────────────────────────
WB_BASE = "https://api.worldbank.org/v2"

def fetch_wb_indicator(indicator, country="WLD", per_page=1000):
    url = f"{WB_BASE}/country/{country}/indicator/{indicator}"
    params = {"format": "json", "per_page": per_page, "mrv": 60}
    try:
        r = requests.get(url, params=params, impersonate=IMPERSONATE, timeout=30)
        r.raise_for_status()
        payload = r.json()
        if len(payload) < 2 or not payload[1]:
            return pd.DataFrame(columns=["year", "value"])
        rows = [{"year": int(d["date"]), "value": d["value"]}
                for d in payload[1] if d["value"] is not None]
        return pd.DataFrame(rows).sort_values("year").reset_index(drop=True)
    except Exception as e:
        print(f"  World Bank fetch failed for {indicator}: {e}")
        return pd.DataFrame(columns=["year", "value"])


def to_month_end(df, date_col="date"):
    df = df.copy()
    df[date_col] = pd.to_datetime(df[date_col]).dt.to_period("M").dt.to_timestamp("M")
    return df

def ffill_annual_to_monthly(df, value_col, year_col="year"):
    df = df.copy()
    df["date"] = pd.to_datetime(df[year_col].astype(str) + "-12-31").dt.to_period("M").dt.to_timestamp("M")
    s = df.set_index("date")[value_col]
    monthly_idx = pd.date_range(s.index.min(), s.index.max() + pd.DateOffset(months=11), freq="ME")
    return s.reindex(monthly_idx).ffill().reset_index().rename(columns={"index": "date"})

print("\u2705 curl_cffi loaded \u2014 Chrome TLS impersonation active")
print(f"\U0001f4c5 Date range: {START_YEAR} \u2192 {END_DATE}")
```

  <h2 style="color:#f0fdf4; margin:4px 0 0; font-size:18px;">🌿 FAO Food Price Index — Overall &amp; 5 Sub-Indices</h2>
  <p style="color:#4ade80; margin:4px 0 0; font-size:13px;">
    Downloads the FAOSTAT Food Price Indices bulk ZIP, filters to the six headline series, pivots to
    a wide monthly panel, and computes month-over-month and year-over-year change rates for each index.<br><br>
    • <strong style="color:#bbf7d0;">FFPI Overall</strong>: FAO Food Price Index — the benchmark global food price tracker (monthly since Jan 1990, base 2014–2016 = 100)<br>
    • <strong style="color:#bbf7d0;">Cereals</strong>: Wheat, maize, rice, barley, sorghum export price quotes — the largest hunger driver<br>
    • <strong style="color:#bbf7d0;">Dairy</strong>: Butter, SMP, WMP, cheese — reflects protein &amp; cold-chain costs<br>
    • <strong style="color:#bbf7d0;">Meat</strong>: Bovine, poultry, ovine, pigmeat export prices<br>
    • <strong style="color:#bbf7d0;">Vegetable Oils</strong>: Palm, soy, rapeseed, sunflower oil — critical for cooking fuel costs in developing markets<br>
    • <strong style="color:#bbf7d0;">Sugar</strong>: ISO daily sugar price quotes, reweighted to monthly average<br>
    Saves to <code>fao_food_prices.csv</code>.
  </p>
</div>

```python
# FAOSTAT item codes for the 6 Food Price Indices
FFPI_ITEMS = {
    23013: 'ffpi_overall',
    23014: 'ffpi_meat',
    23015: 'ffpi_dairy',
    23016: 'ffpi_cereals',
    23017: 'ffpi_oils',
    23018: 'ffpi_sugar',
}

print('Step 1: Fetching FAO Food Price Indices...')
raw_fppi = fetch_faostat_zip('FPPI_E_All_Data_(Normalized).zip')
```

```python
# ── Parse and pivot FFPI data ─────────────────────────────────────────────
if raw_fppi.empty:
    print('⚠️  FFPI download failed — creating empty placeholder')
    fao_prices_df = pd.DataFrame(columns=['date'] + list(FFPI_ITEMS.values()))
else:
    # Normalise column names (FAOSTAT uses 'Item Code' or 'Item Code (FBS)' depending on vintage)
    raw_fppi.columns = raw_fppi.columns.str.strip()
    item_col = 'Item Code' if 'Item Code' in raw_fppi.columns else raw_fppi.columns[2]
    year_col = 'Year'
    val_col  = 'Value'

    # FFPI is stored with months encoded in the Year column as YYYY-MM or
    # in separate Month columns depending on the ZIP vintage
    # Handle both: wide (year columns Y1990M01 … ) and long (Year, Months) formats
    if 'Months' in raw_fppi.columns:
        # Long format with explicit Month column
        month_map = {
            'January': 1, 'February': 2, 'March': 3, 'April': 4,
            'May': 5, 'June': 6, 'July': 7, 'August': 8,
            'September': 9, 'October': 10, 'November': 11, 'December': 12
        }
        raw_fppi['month_num'] = raw_fppi['Months'].map(month_map)
        raw_fppi['date'] = pd.to_datetime(
            raw_fppi['Year'].astype(str) + '-' +
            raw_fppi['month_num'].astype(str).str.zfill(2) + '-01'
        ).dt.to_period('M').dt.to_timestamp('M')
    else:
        # Year column contains YYYYMM integer (e.g. 199001)
        raw_fppi['date'] = pd.to_datetime(
            raw_fppi[year_col].astype(str).str[:4] + '-' +
            raw_fppi[year_col].astype(str).str[4:6].str.zfill(2) + '-01',
            errors='coerce'
        ).dt.to_period('M').dt.to_timestamp('M')

    # Filter to FFPI items and pivot wide
    fppi_long = raw_fppi[raw_fppi[item_col].isin(FFPI_ITEMS.keys())].copy()
    fppi_long['series'] = fppi_long[item_col].map(FFPI_ITEMS)
    fppi_long[val_col] = pd.to_numeric(fppi_long[val_col], errors='coerce')

    fao_prices_df = (
        fppi_long
        .dropna(subset=['date', val_col])
        .groupby(['date', 'series'], as_index=False)[val_col].mean()
        .pivot(index='date', columns='series', values=val_col)
        .reset_index()
    )
    fao_prices_df.columns.name = None
    fao_prices_df = fao_prices_df.sort_values('date').reset_index(drop=True)

    # Filter to START_YEAR onwards
    fao_prices_df = fao_prices_df[fao_prices_df['date'].dt.year >= START_YEAR].reset_index(drop=True)

    # MoM and YoY change rates for each sub-index
    for col in list(FFPI_ITEMS.values()):
        if col in fao_prices_df.columns:
            fao_prices_df[f'{col}_mom_pct'] = (fao_prices_df[col].pct_change(1) * 100).round(3)
            fao_prices_df[f'{col}_yoy_pct'] = (fao_prices_df[col].pct_change(12) * 100).round(3)

    fao_prices_df['date'] = fao_prices_df['date'].dt.strftime('%Y-%m-%d')
    fao_prices_df.to_csv(os.path.join(OUTPUT_DIR, 'fao_food_prices.csv'), index=False)

    print(f'\nFAO Food Price Indices: {fao_prices_df.shape}')
    print(f'  Date range : {fao_prices_df["date"].min()} → {fao_prices_df["date"].max()}')
    print(f'  Columns    : {list(fao_prices_df.columns)}')
    print()
    for col in list(FFPI_ITEMS.values()):
        if col in fao_prices_df.columns:
            s = fao_prices_df[col].dropna()
            print(f'  {col:<18}  min={s.min():.1f}  max={s.max():.1f}  latest={s.iloc[-1]:.1f}')
    print('\n  fao_food_prices.csv saved ✅')
```

  <p style="color:#4ade80; margin:4px 0 0; font-size:13px;">
    Downloads the World Bank monthly commodity price Pink Sheet (Excel), extracts food commodity spot
    prices and the three critical agricultural fertiliser inputs, and computes fertiliser-to-grain
    price ratios that signal farmer input cost pressure and future food supply risk.<br><br>
    • <strong style="color:#bbf7d0;">Wheat (US HRW)</strong>: Hard Red Winter, US Gulf FOB $/mt — global benchmark grain price<br>
    • <strong style="color:#bbf7d0;">Maize (US #2 Yellow)</strong>: US Gulf FOB $/mt — primary feed grain and ethanol feedstock<br>
    • <strong style="color:#bbf7d0;">Rice (Thailand 25%)</strong>: Bangkok FOB $/mt — staple for 3 billion people in Asia<br>

```python
PINK_SHEET_URL = (
    "https://thedocs.worldbank.org/en/doc/"
    "5d903e848db1d1b83e0ec8f744e55570-0350012021/related/"
    "CMO-Historical-Data-Monthly.xlsx"
)
PINK_SHEET_URL_ALT = (
    "https://www.worldbank.org/content/dam/sites/prospects/"
    "media/images/CMO-Historical-Data-Monthly.xlsx"
)

COMMODITY_MAP = {
    "Wheat, US HRW":         "wheat_usd_mt",
    "Maize":                 "maize_usd_mt",
    "Rice, Thailand, 25%":   "rice_usd_mt",
    "Soybean Oil":           "soybean_oil_usd_mt",
    "DAP":                   "dap_fertiliser_usd_mt",
    "Urea, E. Europe, bulk": "urea_usd_mt",
    "Potassium Chloride":    "mop_potash_usd_mt",
}

print("Step 2: Fetching World Bank Pink Sheet commodity prices...")
pink_raw = pd.DataFrame()

for _url in [PINK_SHEET_URL, PINK_SHEET_URL_ALT]:
    try:
        r = requests.get(_url, impersonate=IMPERSONATE, timeout=120)
        if r.status_code == 200:
            pink_raw = pd.read_excel(
                io.BytesIO(r.content),
                sheet_name="Monthly Prices",
                header=4,
                index_col=0,
            )
            print(f"  \u2705  Pink Sheet loaded  \u2192  {pink_raw.shape[0]} months x {pink_raw.shape[1]} commodities")
            break
        print(f"  \u26a0\ufe0f  {r.status_code} on {_url}")
    except Exception as e:
        print(f"  \u26a0\ufe0f  {e}")

if pink_raw.empty:
    print("  \u274c  Pink Sheet unavailable \u2014 commodity_df will be empty")
```

```python
if pink_raw.empty:
    print('⚠️  Pink Sheet unavailable — creating empty placeholder')
    commodity_df = pd.DataFrame(columns=['date'] + list(COMMODITY_MAP.values()))
else:
    # Parse the date index (Pink Sheet uses '1960M01', '1960M02', … format)
    pink_raw.index = pd.to_datetime(
        pink_raw.index.astype(str)
        .str.replace('M', '-', regex=False)
        .str.strip(),
        errors='coerce'
    ).to_period('M').to_timestamp('M')
    pink_raw = pink_raw[~pink_raw.index.isna()].sort_index()

    # Match column names to our map (flexible substring matching)
    col_rename = {}
    for raw_col in pink_raw.columns:
        raw_str = str(raw_col).strip()
        for key, clean in COMMODITY_MAP.items():
            if key.lower() in raw_str.lower():
                col_rename[raw_col] = clean
                break

    # Extract the commodities we want
    keep_cols = list(col_rename.keys())
    commodity_df = pink_raw[keep_cols].rename(columns=col_rename).reset_index()
    commodity_df.columns = ['date'] + list(commodity_df.columns[1:])
    commodity_df = commodity_df.sort_values('date').reset_index(drop=True)

    # Convert to numeric
    for col in commodity_df.columns[1:]:
        commodity_df[col] = pd.to_numeric(commodity_df[col], errors='coerce')

    # MoM and YoY change rates for each commodity
    for col in list(COMMODITY_MAP.values()):
        if col in commodity_df.columns:
            commodity_df[f'{col}_mom_pct'] = (commodity_df[col].pct_change(1) * 100).round(3)
            commodity_df[f'{col}_yoy_pct'] = (commodity_df[col].pct_change(12) * 100).round(3)

    # Fertiliser-to-grain price ratios (key input cost signal)
    if {'dap_fertiliser_usd_mt', 'wheat_usd_mt'}.issubset(commodity_df.columns):
        commodity_df['dap_to_wheat_ratio'] = (
            commodity_df['dap_fertiliser_usd_mt'] / commodity_df['wheat_usd_mt']
        ).round(3)

    if {'urea_usd_mt', 'maize_usd_mt'}.issubset(commodity_df.columns):
        commodity_df['urea_to_maize_ratio'] = (
            commodity_df['urea_usd_mt'] / commodity_df['maize_usd_mt']
        ).round(3)

    # Composite fertiliser price index (equal-weight average of DAP + Urea + MOP)
    fert_cols = [c for c in ['dap_fertiliser_usd_mt', 'urea_usd_mt', 'mop_potash_usd_mt']
                 if c in commodity_df.columns]
    if fert_cols:
        commodity_df['fertiliser_composite_idx'] = commodity_df[fert_cols].mean(axis=1).round(2)

    commodity_df['date'] = commodity_df['date'].dt.strftime('%Y-%m-%d')
    commodity_df.to_csv(os.path.join(OUTPUT_DIR, 'commodity_prices.csv'), index=False)

    print(f'Commodity prices: {commodity_df.shape}')
    print(f'  Date range : {commodity_df["date"].min()} → {commodity_df["date"].max()}')
    print(f'  Columns    : {list(commodity_df.columns)}')
    print()
    for col in list(COMMODITY_MAP.values()):
        if col in commodity_df.columns:
            s = commodity_df[col].dropna()
            if len(s):
                print(f'  {col:<28}  min={s.min():.1f}  max={s.max():.1f}  latest={s.iloc[-1]:.1f}')
    print('\n  commodity_prices.csv saved ✅')
```

  <span style="color:#86efac; font-size:12px; font-weight:600; letter-spacing:1px;">STEP 3 / 5</span>
  <h2 style="color:#f0fdf4; margin:4px 0 0; font-size:18px;">🌽 Crop Production &amp; Yields — FAOSTAT (Wheat, Rice, Maize)</h2>
  <p style="color:#4ade80; margin:4px 0 0; font-size:13px;">
    Downloads the FAOSTAT crops production bulk ZIP, filters to three staple grains at world-aggregate
    level, and computes area-harvested, production quantity, and yield metrics with year-over-year
    change rates and trend context.<br><br>
    • <strong style="color:#bbf7d0;">Wheat (Item 15)</strong>: Global area (Mha), production (Mt), yield (t/ha) — Jan 1961 to latest FAO estimate<br>
    • <strong style="color:#bbf7d0;">Rice, Paddy (Item 27)</strong>: Global milled-equivalent production — the calorie staple for East &amp; South Asia<br>
    • <strong style="color:#bbf7d0;">Maize (Item 56)</strong>: Global production — driven by feed demand and increasingly biofuel mandates<br>
    • <strong style="color:#bbf7d0;">Elements</strong>: 5312 (Area Harvested, kha), 5419 (Yield, hg/ha → converted to t/ha), 5510 (Production, kt → converted to Mt)<br>
    Saves to <code>crop_yields.csv</code>.
  </p>
</div>

```python
# FAOSTAT crop items and elements we need
CROP_ITEMS = {
    15:  'wheat',
    27:  'rice_paddy',
    56:  'maize',
}

CROP_ELEMENTS = {
    5312: 'area_kha',        # Area Harvested in 1000 ha
    5510: 'production_kt',   # Production in 1000 tonnes
    5419: 'yield_hg_ha',     # Yield in hg/ha (hectogram per hectare)
}

# World aggregate area code = 5000
WORLD_CODE = 5000

print('Step 3: Fetching FAOSTAT crop production data...')
raw_qcl = fetch_faostat_zip('Production_Crops_Livestock_E_All_Data_(Normalized).zip')
```

```python
if raw_qcl.empty:
    print('⚠️  QCL download failed — creating empty placeholder')
    crop_yields_df = pd.DataFrame()
else:
    raw_qcl.columns = raw_qcl.columns.str.strip()
    area_col = 'Area Code' if 'Area Code' in raw_qcl.columns else raw_qcl.columns[0]
    item_col = 'Item Code' if 'Item Code' in raw_qcl.columns else raw_qcl.columns[2]
    elem_col = 'Element Code' if 'Element Code' in raw_qcl.columns else raw_qcl.columns[4]

    # Filter: world aggregate + our 3 crops + our 3 elements
    qcl_f = raw_qcl[
        (raw_qcl[area_col] == WORLD_CODE) &
        (raw_qcl[item_col].isin(CROP_ITEMS.keys())) &
        (raw_qcl[elem_col].isin(CROP_ELEMENTS.keys()))
    ].copy()

    qcl_f['crop']    = qcl_f[item_col].map(CROP_ITEMS)
    qcl_f['element'] = qcl_f[elem_col].map(CROP_ELEMENTS)
    qcl_f['Value']   = pd.to_numeric(qcl_f['Value'], errors='coerce')
    qcl_f['year']    = pd.to_numeric(qcl_f['Year'], errors='coerce').astype('Int64')

    # Pivot to wide: one row per (year, crop), columns for each element
    crop_wide = (
        qcl_f
        .dropna(subset=['Value', 'year'])
        .groupby(['year', 'crop', 'element'], as_index=False)['Value'].mean()
        .pivot_table(index=['year', 'crop'], columns='element', values='Value')
        .reset_index()
    )
    crop_wide.columns.name = None

    # Unit conversions
    if 'area_kha' in crop_wide.columns:
        crop_wide['area_mha'] = (crop_wide['area_kha'] / 1_000).round(3)       # Mha
    if 'production_kt' in crop_wide.columns:
        crop_wide['production_mt'] = (crop_wide['production_kt'] / 1_000).round(3)  # Mt
    if 'yield_hg_ha' in crop_wide.columns:
        crop_wide['yield_t_ha'] = (crop_wide['yield_hg_ha'] / 10_000).round(4)  # t/ha

    # Re-pivot to fully wide: one row per year, columns prefixed by crop
    METRIC_COLS = ['area_mha', 'production_mt', 'yield_t_ha']
    crops_list = []
    for crop, gdf in crop_wide.groupby('crop'):
        available = ['year'] + [c for c in METRIC_COLS if c in gdf.columns]
        gdf = gdf[available].copy()
        gdf.columns = ['year'] + [f'{crop}_{c}' for c in available[1:]]
        crops_list.append(gdf.set_index('year'))

    crop_yields_df = pd.concat(crops_list, axis=1).reset_index()
    crop_yields_df = crop_yields_df.sort_values('year').reset_index(drop=True)

    # YoY change rates for yield and production
    for crop in CROP_ITEMS.values():
        for metric in ['production_mt', 'yield_t_ha']:
            col = f'{crop}_{metric}'
            if col in crop_yields_df.columns:
                crop_yields_df[f'{col}_yoy_pct'] = (
                    crop_yields_df[col].pct_change(1) * 100
                ).round(3)

    # Total staple grain production (wheat + rice + maize)
    prod_cols = [f'{c}_production_mt' for c in CROP_ITEMS.values()
                 if f'{c}_production_mt' in crop_yields_df.columns]
    if prod_cols:
        crop_yields_df['total_staple_production_mt'] = crop_yields_df[prod_cols].sum(axis=1).round(3)

    crop_yields_df.to_csv(os.path.join(OUTPUT_DIR, 'crop_yields.csv'), index=False)

    print(f'Crop yields: {crop_yields_df.shape}')
    print(f'  Year range : {int(crop_yields_df["year"].min())} → {int(crop_yields_df["year"].max())}')
    print(f'  Columns    : {list(crop_yields_df.columns)}')
    print()
    for crop in CROP_ITEMS.values():
        ycol = f'{crop}_yield_t_ha'
        pcol = f'{crop}_production_mt'
        if ycol in crop_yields_df.columns:
            y = crop_yields_df[ycol].dropna()
            p = crop_yields_df[pcol].dropna()
            print(f'  {crop:<14}  yield {y.iloc[0]:.2f}→{y.iloc[-1]:.2f} t/ha  '
                  f'production {p.iloc[-1]:.0f} Mt (latest)')
    print('\n  crop_yields.csv saved ✅')
```

<div style="border-radius:10px; background:#1c2c1c; padding:16px 22px; margin:8px 0;">
  <span style="color:#86efac; font-size:12px; font-weight:600; letter-spacing:1px;">STEP 4 / 5</span>
  <h2 style="color:#f0fdf4; margin:4px 0 0; font-size:18px;">🍽️ Food Security Indicators — World Bank API</h2>
  <p style="color:#4ade80; margin:4px 0 0; font-size:13px;">
    Fetches five annual World Bank indicators covering undernourishment, food production capacity, and
    agricultural productivity. Annual data is forward-filled to a monthly grid for panel merging.<br><br>
    • <strong style="color:#bbf7d0;">SN.ITK.DEFC.ZS</strong>: Prevalence of undernourishment (% of population) — the headline hunger rate (annual, 1992)<br>
    • <strong style="color:#bbf7d0;">SN.ITK.MSFI.ZS</strong>: Prevalence of moderate or severe food insecurity (% adults) — FAO FIES survey-based (2015)<br>
    • <strong style="color:#bbf7d0;">AG.PRD.FOOD.XD</strong>: Food production index (2014–2016 = 100) — tracks global agricultural output capacity (1961)<br>
    • <strong style="color:#bbf7d0;">AG.YLD.CREL.KG</strong>: Cereal yield (kg/ha) — aggregate of all cereals, global average (1961)<br>
    • <strong style="color:#bbf7d0;">AG.LND.ARBL.ZS</strong>: Arable land as % of land area — long-run supply constraint indicator (1961)<br>
    Saves to <code>food_security_indicators.csv</code>.
  </p>
</div>

```python
WB_FOOD_INDICATORS = {
    'SN.ITK.DEFC.ZS': 'undernourishment_pct',          # % population, annual
    'SN.ITK.MSFI.ZS': 'food_insecurity_moderate_pct',  # % adults, annual (from 2015)
    'AG.PRD.FOOD.XD':  'food_production_index',         # 2014-2016=100, annual
    'AG.YLD.CREL.KG':  'cereal_yield_kg_ha',            # kg per hectare, annual
    'AG.LND.ARBL.ZS':  'arable_land_pct',               # % land area, annual
}

print('Step 4: Fetching World Bank food security indicators...')
wb_annual = {}

for indicator, col in WB_FOOD_INDICATORS.items():
    df = fetch_wb_indicator(indicator, country='WLD', per_page=200)
    if df.empty:
        print(f'  {indicator}: no data returned')
        continue
    df = df.rename(columns={'value': col})
    wb_annual[col] = df.set_index('year')[col]
    ymin, ymax = df['year'].min(), df['year'].max()
    latest = df.loc[df['year'].idxmax(), col]
    print(f'  {indicator:<22}  {col:<35}  {len(df):>3} years  [{ymin}→{ymax}]  latest={latest:.2f}')
    time.sleep(0.5)

# Merge into a single annual DataFrame
if wb_annual:
    fs_annual = pd.DataFrame(wb_annual).reset_index().rename(columns={'index': 'year'})
    fs_annual = fs_annual.sort_values('year').reset_index(drop=True)

    # YoY change rates
    for col in ['undernourishment_pct', 'food_production_index', 'cereal_yield_kg_ha']:
        if col in fs_annual.columns:
            fs_annual[f'{col}_yoy_chg'] = fs_annual[col].diff(1).round(3)

    fs_annual.to_csv(os.path.join(OUTPUT_DIR, 'food_security_indicators.csv'), index=False)

    print(f'\nFood security indicators: {fs_annual.shape}')
    print(f'  Year range : {int(fs_annual["year"].min())} → {int(fs_annual["year"].max())}')
    print(f'  Columns    : {list(fs_annual.columns)}')
    print('  food_security_indicators.csv saved ✅')
else:
    print('⚠️  No World Bank data fetched — creating empty placeholder')
    fs_annual = pd.DataFrame(columns=['year'] + list(WB_FOOD_INDICATORS.values()))
```

<div style="border-radius:10px; background:#1c2c1c; padding:16px 22px; margin:8px 0;">
  <span style="color:#86efac; font-size:12px; font-weight:600; letter-spacing:1px;">STEP 5 / 5</span>
  <h2 style="color:#f0fdf4; margin:4px 0 0; font-size:18px;">🔗 Food Security Panel + ML Feature Engineering</h2>
  <p style="color:#4ade80; margin:4px 0 0; font-size:13px;">
    Merges all four source files onto a monthly date spine, computes the six derived food security ratios,
    then engineers the ML feature matrix with lag features, rolling windows, crisis flags, and lead targets.
    Designed for cross-reference with tariff, trade, and macroeconomic datasets.<br><br>
    • <strong style="color:#bbf7d0;">Real Food Price Index</strong>: Nominal FFPI deflated by world CPI — true purchasing-power cost of food<br>
    • <strong style="color:#bbf7d0;">Fertiliser-to-Wheat Ratio</strong>: DAP price ÷ wheat price — signals when input costs crush farmer margins<br>
    • <strong style="color:#bbf7d0;">Food Price Volatility</strong>: 12-month rolling σ of MoM FFPI change — identifies shock regimes (2008, 2011, 2022)<br>
    • <strong style="color:#bbf7d0;">Food vs Headline CPI Gap</strong>: FFPI YoY − World CPI YoY — disproportionate food inflation signal<br>
    • <strong style="color:#bbf7d0;">Cereals Affordability Index</strong>: Inverse of cereals sub-index, rebased to 100 — below 100 = worse than baseline<br>
    • <strong style="color:#bbf7d0;">Supply Stress Flag</strong>: Binary — simultaneous 20%+ spike in cereals prices AND fertiliser costs<br>
    • <strong style="color:#bbf7d0;">ML Targets</strong>: Next-month FFPI % change (regression), 12-month direction (classification), next-month cereals change
  </p>
</div>

```python
# ── Forward-fill annual data to monthly grid ───────────────────────────────
def to_dt(df, col='date'):
    df = df.copy()
    df[col] = pd.to_datetime(df[col])
    return df


fao_dt  = to_dt(fao_prices_df) if not fao_prices_df.empty else pd.DataFrame()
comm_dt = to_dt(commodity_df)  if not commodity_df.empty else pd.DataFrame()

# Annual data → monthly forward-fill for panel merge
fs_monthly_parts = []
if not fs_annual.empty and 'year' in fs_annual.columns:
    for col in [c for c in fs_annual.columns if c != 'year']:
        series_df = fs_annual[['year', col]].dropna(subset=[col])
        if series_df.empty:
            continue
        m = ffill_annual_to_monthly(series_df, col, year_col='year')
        m = m.rename(columns={col: col})
        fs_monthly_parts.append(m.set_index('date')[col])
    if fs_monthly_parts:
        fs_monthly = pd.DataFrame(fs_monthly_parts).T.reset_index().rename(columns={'index': 'date'})
    else:
        fs_monthly = pd.DataFrame(columns=['date'])
else:
    fs_monthly = pd.DataFrame(columns=['date'])

print('Annual food security indicators forward-filled to monthly ✅')
if not fs_monthly.empty and 'date' in fs_monthly.columns:
    print(f'  Monthly grid: {fs_monthly.shape}')
```

```python
# ── Build monthly panel ────────────────────────────────────────────────────
# Use FAO food prices as the spine (monthly since 1990)
if fao_dt.empty:
    print('⚠️  FAO price data unavailable — panel will be sparse')
    panel = comm_dt.copy() if not comm_dt.empty else pd.DataFrame(columns=['date'])
else:
    panel = fao_dt.copy()

if not comm_dt.empty and 'date' in comm_dt.columns:
    panel = panel.merge(comm_dt, on='date', how='outer')

if not fs_monthly.empty and 'date' in fs_monthly.columns:
    panel = panel.merge(fs_monthly, on='date', how='left')

panel = panel.sort_values('date').reset_index(drop=True)

# Trim to START_YEAR
panel = panel[panel['date'].dt.year >= START_YEAR].reset_index(drop=True)

print(f'Panel shape after merge: {panel.shape}')
```

```python
# ── Derived Food Security Ratios ──────────────────────────────────────────

# 1. Real Food Price Index — deflate FFPI by World CPI
# We use the FFPI overall as the nominal index and derive a real version
# World CPI is fetched separately via World Bank if available
# For now we use FFPI as-is (already real-ish; FAO rebases to 2014-2016=100)
# Flag if FFPI > 120 as an elevated price regime
if 'ffpi_overall' in panel.columns:
    panel['ffpi_above_baseline'] = (panel['ffpi_overall'] > 100).astype(int)
    panel['ffpi_crisis_flag']    = (panel['ffpi_overall'] > 130).astype(int)  # 2011 / 2022 levels

# 2. Fertiliser-to-Wheat Price Ratio
if {'dap_fertiliser_usd_mt', 'wheat_usd_mt'}.issubset(panel.columns):
    panel['dap_to_wheat_ratio'] = (
        panel['dap_fertiliser_usd_mt'] / panel['wheat_usd_mt']
    ).round(3)
    panel['dap_to_wheat_elevated'] = (panel['dap_to_wheat_ratio'] > 3.5).astype(int)

# 3. Food Price Volatility — 12-month rolling σ of MoM FFPI change
if 'ffpi_overall_mom_pct' in panel.columns:
    panel['ffpi_volatility_12m'] = (
        panel['ffpi_overall_mom_pct'].rolling(12).std()
    ).round(3)

# 4. Cereals Affordability Index — inverse of cereals sub-index (base 100 = 2014-2016)
if 'ffpi_cereals' in panel.columns:
    panel['cereals_affordability_idx'] = (
        (100 / panel['ffpi_cereals']) * 100
    ).round(2)

# 5. Supply Stress Flag — cereals YoY > 20% AND fertiliser YoY > 20% simultaneously
if {'ffpi_cereals_yoy_pct', 'dap_fertiliser_usd_mt_yoy_pct'}.issubset(panel.columns):
    panel['supply_stress_flag'] = (
        (panel['ffpi_cereals_yoy_pct'] > 20) &
        (panel['dap_fertiliser_usd_mt_yoy_pct'] > 20)
    ).astype(int)

# 6. Composite food shock score — normalised sum of stress signals
stress_cols = [c for c in [
    'ffpi_crisis_flag', 'dap_to_wheat_elevated', 'supply_stress_flag'
] if c in panel.columns]
if stress_cols:
    panel['food_shock_score'] = panel[stress_cols].sum(axis=1)

# Calendar features
panel['month'] = panel['date'].dt.month
panel['year']  = panel['date'].dt.year

panel['date'] = panel['date'].dt.strftime('%Y-%m-%d')
panel.to_csv(os.path.join(OUTPUT_DIR, 'food_security_panel.csv'), index=False)

print(f'Food security panel: {panel.shape}')
print(f'  Date range : {panel["date"].min()} → {panel["date"].max()}')
print()
for col, label in [
    ('ffpi_overall',              'FFPI Overall          '),
    ('ffpi_cereals',              'FFPI Cereals          '),
    ('cereals_affordability_idx', 'Cereals Afford. Idx   '),
    ('dap_to_wheat_ratio',        'DAP/Wheat Ratio       '),
    ('ffpi_volatility_12m',       'FFPI Volatility (12m) '),
]:
    if col in panel.columns:
        s = panel[col].dropna()
        if len(s):
            print(f'  {label}  min={s.min():.2f}  max={s.max():.2f}  latest={s.iloc[-1]:.2f}')

if 'supply_stress_flag' in panel.columns:
    stress_months = panel['supply_stress_flag'].sum()
    print(f'\n  Supply stress months flagged: {stress_months} '
          f'({stress_months / len(panel) * 100:.1f}% of history)')

print('\n  food_security_panel.csv saved ✅')
```

```python
# ── ML Feature Engineering ────────────────────────────────────────────────
ml = panel.copy()
ml['date'] = pd.to_datetime(ml['date'])
ml = ml.sort_values('date').reset_index(drop=True)

# Core series for feature engineering
KEY_SERIES = [c for c in [
    'ffpi_overall', 'ffpi_cereals', 'ffpi_dairy', 'ffpi_meat', 'ffpi_oils', 'ffpi_sugar',
    'wheat_usd_mt', 'maize_usd_mt', 'rice_usd_mt', 'soybean_oil_usd_mt',
    'dap_fertiliser_usd_mt', 'urea_usd_mt', 'mop_potash_usd_mt', 'fertiliser_composite_idx',
    'dap_to_wheat_ratio', 'cereals_affordability_idx', 'ffpi_volatility_12m',
    'undernourishment_pct', 'food_production_index', 'cereal_yield_kg_ha',
] if c in ml.columns]

# ── Lag features L1 – L6 ─────────────────────────────────────────────────
for lag in range(1, 7):
    for col in KEY_SERIES:
        ml[f'{col}_L{lag}'] = ml[col].shift(lag)

# ── Rolling 3m / 6m / 12m mean and std ───────────────────────────────────
ROLL_COLS = [c for c in [
    'ffpi_overall', 'ffpi_cereals', 'wheat_usd_mt', 'maize_usd_mt',
    'dap_fertiliser_usd_mt', 'fertiliser_composite_idx',
    'cereals_affordability_idx', 'dap_to_wheat_ratio',
] if c in ml.columns]

for window in [3, 6, 12]:
    for col in ROLL_COLS:
        ml[f'{col}_roll_mean_{window}m'] = ml[col].rolling(window).mean().round(4)
        ml[f'{col}_roll_std_{window}m']  = ml[col].rolling(window).std().round(4)

# ── MoM and YoY differencing ─────────────────────────────────────────────
for col in ['ffpi_overall', 'ffpi_cereals', 'wheat_usd_mt',
            'dap_fertiliser_usd_mt', 'fertiliser_composite_idx']:
    if col in ml.columns:
        ml[f'{col}_mom'] = ml[col].diff(1).round(4)
        ml[f'{col}_yoy'] = ml[col].diff(12).round(4)

# ── Interaction features ──────────────────────────────────────────────────
# Fertiliser × grain price: compound cost-push shock signal
if {'fertiliser_composite_idx', 'ffpi_cereals'}.issubset(ml.columns):
    ml['fert_x_cereals'] = (
        ml['fertiliser_composite_idx'] * ml['ffpi_cereals'] / 1e4
    ).round(4)

# Demand pressure: food price rising while undernourishment is also rising
if {'ffpi_overall', 'undernourishment_pct'}.issubset(ml.columns):
    ml['ffpi_x_hunger'] = (
        ml['ffpi_overall'] * ml['undernourishment_pct'] / 1e3
    ).round(4)

# ── Regime features ───────────────────────────────────────────────────────
if 'ffpi_overall' in ml.columns:
    ml['ffpi_regime'] = pd.cut(
        ml['ffpi_overall'],
        bins=[0, 80, 100, 120, 999],
        labels=['deflation', 'normal', 'elevated', 'crisis']
    ).astype(str)
    ml['ffpi_regime_code'] = ml['ffpi_regime'].map(
        {'deflation': 0, 'normal': 1, 'elevated': 2, 'crisis': 3}
    )

# ── Targets ──────────────────────────────────────────────────────────────
# Regression: next-month and next-12-month FFPI percent change
if 'ffpi_overall' in ml.columns:
    ml['target_ffpi_chg_1m']  = (ml['ffpi_overall'].pct_change(1)  * 100).shift(-1).round(4)
    ml['target_ffpi_chg_12m'] = (ml['ffpi_overall'].pct_change(12) * 100).shift(-12).round(4)

# Classification: will FFPI be higher 12 months from now?
if 'target_ffpi_chg_12m' in ml.columns:
    ml['target_ffpi_up_12m'] = (ml['target_ffpi_chg_12m'] > 0).astype(float)

# Regression: next-month cereals price change
if 'ffpi_cereals' in ml.columns:
    ml['target_cereals_chg_1m'] = ml['ffpi_cereals'].diff(1).shift(-1).round(3)

# Crisis prediction: will food shock score be >= 2 next month?
if 'food_shock_score' in ml.columns:
    ml['target_shock_next_month'] = (ml['food_shock_score'].shift(-1) >= 2).astype(float)

# ── Final cleanup ─────────────────────────────────────────────────────────
ml = ml.dropna(subset=['target_ffpi_chg_1m'])
ml['date'] = ml['date'].dt.strftime('%Y-%m-%d')
ml.to_csv(os.path.join(OUTPUT_DIR, 'ml_features.csv'), index=False)

print(f'ML feature matrix: {ml.shape}')
print(f'  Date range   : {ml["date"].min()} → {ml["date"].max()}')
print(f'  Total cols   : {ml.shape[1]}')
print(f'  Targets      : target_ffpi_chg_1m | target_ffpi_chg_12m | target_ffpi_up_12m | target_cereals_chg_1m | target_shock_next_month')
if 'target_ffpi_up_12m' in ml.columns:
    up_rate = ml['target_ffpi_up_12m'].mean()
    print(f'  FFPI-up rate : {up_rate*100:.1f}% of months (classification base rate)')
print('  ml_features.csv saved ✅')
```
